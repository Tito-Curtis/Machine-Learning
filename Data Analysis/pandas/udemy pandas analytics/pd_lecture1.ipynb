```python
## oop concept in visualizaton
fig,ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter(
                x=over_50['age'],
                y=over_50['chol'],
                c=over_50['target'])
ax.set(title="Hear Diseases in ghana",
       xlabel='Age/(yrs)',
       ylabel='Cholestrol Level');
ax.legend(*scatter.legend_elements(), title="Target")
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
over_50.head()
```

```python
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize = (10,10))
ax0.scatter(x=over_50["age"],
            y=over_50["chol"],
            c=over_50["target"])
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.set(title="Heart diseases (Age and choleterol)",xlabel="Age", ylabel="Cholesterol")
ax0.axhline(over_50['chol'].mean(), linestyle="--");
ax1.scatter(x=over_50["age"],
            y=over_50["thalach"],
            c=over_50["target"])
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.set(title="Heart diseases (Age and Maximum Heart Beat)",xlabel="Age", ylabel="Thalach")
ax1.axhline(over_50['thalach'].mean(), linestyle="--")
fig.suptitle("Heart Diseases Analysis", fontsize=16, fontweight="bold");
```
