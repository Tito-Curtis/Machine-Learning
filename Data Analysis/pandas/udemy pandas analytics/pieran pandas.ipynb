## merging DataFrame 

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'names':['Andrew','Bobo','Claire','David']})
```

```python
logins = pd.DataFrame({'log_id':[1,2,3,4],'names':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
pd.merge(registrations,logins,how='inner',on='names')
```

```python
pd.merge(registrations,logins,how='outer',on='names')
```

```python
pd.merge(left=registrations, right=logins,how='left',on='names')
```

```python
pd.merge(registrations,logins,how='outer',on='names')
```
