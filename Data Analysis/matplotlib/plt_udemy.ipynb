```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot()
plt.show()
```

```python
x=[1,2,3,4]
y=[11,22,33,44]
plt.plot(x,y)
plt.show()
```

```python
#object oriented pyplot, 1st method
fig = plt.figure() # add figure
ax =fig.add_subplot() #add axes
plt.show()
```

```python
#method 2
fig= plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
 
```

```python
#method 3
fig, ax = plt.subplots()
ax.plot(x,y);
```

 ## Matplotlib examples workflow

```python
x =[1,2,3,4]
y =[11,22,33,44]
fig,ax=plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title="Simple Plot", xlabel="x-axis",ylabel="y-axis")
fig.savefig("images/Sample-plot.png")
```

## Making figures with Numpy array

```python
x = np.linspace(0,10,100)
x[:10]
```

```python
#line plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
ax.set(title="Line plot"):
```

```python
# using same data to make a scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# another scatter plot
fig,ax =plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#Making a plot from a dictionary
nut_butter_prices = {
    "Almond butter":10,
    "Peanut butter":8,
    "Cashew butter":12
}
fig,ax=plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title="Butter Prices in UCC",ylabel="Prices($)",xlabel="Butter Name")
```

```python
fig,ax=plt.subplots()
ax.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()));
ax.set(title="Butter Prices in UCC",xlabel="Prices($)",ylabel="Butter Name");
```

```python
x = np.random.randn(1000)
fig,ax=plt.subplots()
ax.hist(x);
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(x,np.exp(x));
ax3.hist(x);
ax4.set(title="Butter Prices in UCC",xlabel="Prices($)",ylabel="Butter Name");


                                          
```
